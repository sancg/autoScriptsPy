# Use `pathlib` instead of `os.path`   
In Python, modern filesystem work should use pathlib. It is cleaner and safer than os.path.

You also have:
- os.walk() → classic recursive traversal
- pathlib.Path.rglob() → recursive globbing
- pathlib.Path.iterdir() → shallow traversal

For controlled depth and filtering, `os.walk()` is actually very powerful because it lets you modify traversal behavior mid-walk.

```python
# 1. Basic recursive traversal
from pathlib import Path

def walk_all(root):
    root = Path(root)
    print(root)
    for path in root.rglob("*"):
        if path.is_file():
            print(path)
walk_all('../notebooks/')            
```

### What does `os.walk()` that is not useful with `Path("./folder").rglob("*")`
***Controlled Depth Traversal*** - 
Depth is calculated by counting directory separators.

```python
from pathlib import Path

def pathComponents(root):
    root = Path(root).resolve()
    root_depth = len(root.parts)
    print((root, root_depth))
    
pathComponents('../')
```

```python
import os
from pathlib import Path

def walk_with_depth(root, max_depth=None):
    root = Path(root).resolve()
    root_depth = len(root.parts)

    for current_path, dirs, files in os.walk(root):
        current_depth = len(Path(current_path).parts) - root_depth
        if max_depth is not None and current_depth > max_depth:
            # stop descending further
            dirs[:] = []
            continue
        
        for file in files:
            yield Path(current_path) / file

files = walk_with_depth('../', 2)
print(files)
```
